```python
# Importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset

df=pd.read_csv('adspend.csv')
df
```

```python
# Label encoding the features

from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
#df['Survived']=label_encoder.fit_transform(df['Survived'])
#df['Pclass']=label_encoder.fit_transform(df['Pclass'])
#df['Sex']=label_encoder.fit_transform(df['Sex'])
#df['Age']=label_encoder.fit_transform(df['Age'])
#df['SibSp']=label_encoder.fit_transform(df['SibSp'])
#df['Parch']=label_encoder.fit_transform(df['Parch'])
#df['Ticket']=label_encoder.fit_transform(df['Ticket'])
df['date']=label_encoder.fit_transform(df['date'])
df['ad']=label_encoder.fit_transform(df['ad'])

df
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
date_encoder = LabelEncoder()
ad_encoder = LabelEncoder()

# Fit the encoders on specific columns
df['date'] = date_encoder.fit_transform(df['date'])
df['ad'] = ad_encoder.fit_transform(df['ad'])

# Store the mappings for decoding
date_mapping = dict(zip(date_encoder.classes_, date_encoder.transform(date_encoder.classes_)))
ad_mapping = dict(zip(ad_encoder.classes_, ad_encoder.transform(ad_encoder.classes_)))

print("Date Mapping:", date_mapping)
print("Ad Mapping:", ad_mapping)
```

```python
print(df.index)
```

```python
# Converting Index to dataframe

df.index = pd.to_datetime(df.index, errors='coerce')
df.index = pd.to_datetime(df.index, format='%Y-%m-%d %I-%p')
df
```

```python
X = df[['date','shown','converted','avg_cost_per_click','total_revenue','ad']]
y = df[['clicked']]
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
encoder = LabelEncoder()

# Extract the column as a 1D array
y = df['date']  # Ensure it's 1D

# Fit and transform the data
df['date'] = encoder.fit_transform(y)
```

```python
# Split into training and testing data

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
# Initialise the model

clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
# Predictions

y_pred = clf.predict(X_test)
y_pred
```

```python
# Make predictions on the test data
y_pred = clf.predict(X_test)

# Count how many ads were predicted to be clicked
clicked_count = sum(y_pred)  # This gives the number of 1's (clicked)
print(f"Number of ads predicted to be clicked: {clicked_count}")
```

```python
# Print the accuracy of the test data

accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Model Accuracy: {accuracy:.2f}%")
```

```python
# Decode predictions or feature fields for user input

original_date = date_encoder.inverse_transform([1])[0]  # Decode '1' back to its original value
original_ad = ad_encoder.inverse_transform([2])[0]      # Decode '2' back to its original value

print("Decoded Date:", original_date)
print("Decoded Ad:", original_ad)
```

```python
# Example test case after decoding

test_input_decoded = {
    'date': original_date,             # Original date value
    'shown': 300,                      # Ad impressions
    'converted': 20,                   # Conversions
    'avg_cost_per_click': 0.75,        # Cost per click
    'total_revenue': 150.0,            # Revenue
    'ad': original_ad                  # Original ad value
}

# Convert test input to encoded form
test_input_encoded = pd.DataFrame({
    'date': [date_encoder.transform([test_input_decoded['date']])[0]],
    'shown': [test_input_decoded['shown']],
    'converted': [test_input_decoded['converted']],
    'avg_cost_per_click': [test_input_decoded['avg_cost_per_click']],
    'total_revenue': [test_input_decoded['total_revenue']],
    'ad': [ad_encoder.transform([test_input_decoded['ad']])[0]]
})

# Make predictions on the encoded input
test_prediction = clf.predict(test_input_encoded)
print("Test Input (Decoded):", test_input_decoded)
print("Prediction:", test_prediction)
```
