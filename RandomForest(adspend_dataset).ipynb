```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('adspend.csv')
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df['date']=label_encoder.fit_transform(df['date'])
df['ad']=label_encoder.fit_transform(df['ad'])

df
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
date_encoder = LabelEncoder()
ad_encoder = LabelEncoder()

# Fit the encoders on specific columns
df['date'] = date_encoder.fit_transform(df['date'])
df['ad'] = ad_encoder.fit_transform(df['ad'])

# Store the mappings for decoding
date_mapping = dict(zip(date_encoder.classes_, date_encoder.transform(date_encoder.classes_)))
ad_mapping = dict(zip(ad_encoder.classes_, ad_encoder.transform(ad_encoder.classes_)))

print("Date Mapping:", date_mapping)
print("Ad Mapping:", ad_mapping)
```

```python
print(df.index)
```

```python
df.index = pd.to_datetime(df.index, errors='coerce')
```

```python
df.index = pd.to_datetime(df.index, format='%Y-%m-%d %I-%p')
df
```

```python
X = df[['date','shown','converted','avg_cost_per_click','total_revenue','ad']]
y = df[['clicked']]
X
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
encoder = LabelEncoder()

# Extract the column as a 1D array
y = df['date']  # Ensure it's 1D

# Fit and transform the data
df['date'] = encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
# Make predictions on the test data
y_pred = clf.predict(X_test)

# Count how many ads were predicted to be clicked
clicked_count = sum(y_pred)  # This gives the number of 1's (clicked)
print(f"Number of ads predicted to be clicked: {clicked_count}")
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f} %")
```

```python
# Decode predictions or feature fields for user input
original_date = date_encoder.inverse_transform([1])[0]  # Decode '1' back to its original value
original_ad = ad_encoder.inverse_transform([2])[0]      # Decode '2' back to its original value

print("Decoded Date:", original_date)
print("Decoded Ad:", original_ad)
```

```python
# Example test case after decoding
test_input_decoded = {
    'date': original_date,             # Original date value
    'shown': 300,                      # Ad impressions
    'converted': 20,                   # Conversions
    'avg_cost_per_click': 0.75,        # Cost per click
    'total_revenue': 150.0,            # Revenue
    'ad': original_ad                  # Original ad value
}

# Convert test input to encoded form
test_input_encoded = pd.DataFrame({
    'date': [date_encoder.transform([test_input_decoded['date']])[0]],
    'shown': [test_input_decoded['shown']],
    'converted': [test_input_decoded['converted']],
    'avg_cost_per_click': [test_input_decoded['avg_cost_per_click']],
    'total_revenue': [test_input_decoded['total_revenue']],
    'ad': [ad_encoder.transform([test_input_decoded['ad']])[0]]
})

# Make predictions on the encoded input
test_prediction = clf.predict(test_input_encoded)
print("Test Input (Decoded):", test_input_decoded)
print("Prediction:", test_prediction)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Label Encode categorical features
date_encoder = LabelEncoder()
ad_encoder = LabelEncoder()

df['date'] = date_encoder.fit_transform(df['date'])
df['ad'] = ad_encoder.fit_transform(df['ad'])

# --------------------------
# Predicting avg_cost_per_click
# --------------------------
X_click = df.drop(columns=['avg_cost_per_click', 'total_revenue'])
y_click = df['avg_cost_per_click']

X_train_click, X_test_click, y_train_click, y_test_click = train_test_split(X_click, y_click, test_size=0.25)

model_click = RandomForestRegressor()
model_click.fit(X_train_click, y_train_click)

y_pred_click = model_click.predict(X_test_click)

rmse_click = np.sqrt(mean_squared_error(y_test_click, y_pred_click))
r2_click = r2_score(y_test_click, y_pred_click)

print("----- Avg_cost_per_click -----")
print(f"RMSE: {rmse_click:.2f}")
print(f"RÂ² Score: {r2_click:.2f}\n")

# --------------------------
# Predicting total_revenue
# --------------------------
X_revenue = df.drop(columns=['avg_cost_per_click', 'total_revenue'])
y_revenue = df['total_revenue']

X_train_rev, X_test_rev, y_train_rev, y_test_rev = train_test_split(X_revenue, y_revenue, test_size=0.25)

model_rev = RandomForestRegressor()
model_rev.fit(X_train_rev, y_train_rev)

y_pred_rev = model_rev.predict(X_test_rev)

rmse_rev = np.sqrt(mean_squared_error(y_test_rev, y_pred_rev))
r2_rev = r2_score(y_test_rev, y_pred_rev)

print("----- Total_revenue -----")
print(f"RMSE: {rmse_rev:.2f}")
print(f"RÂ² Score: {r2_rev:.2f}")
```
